```python
# Firstly we create the DataFrame with the months groupped per time period
```

```python
# finding each row from column "Date" that starts with "Ma"
ma = ex6.iloc[:,0].str.startswith("Ma")
# creating a DataFrame containing only rows that start with "Ma" from all columns
ex6_filt = ex6[ma]
# finding the total sum of column "Earnings on investments"
value_sum = ex6_filt.iloc[:,1].sum()
value_sum
```

```python
def add_row_with_column_sums(df):
    # find the sum of each column
    col_sums = df.sum()
    # create a new DataFrame with the column sums as the only row
    new_row = pd.DataFrame([col_sums], columns=df.columns)
    # append the new row to the original DataFrame and reset the index
    df_with_total = df.append(new_row, ignore_index=True)
    return df_with_total

# adding a new row with the use of the function
df_with_total= add_row_with_column_sums(ex6_filt)
# dropping all indexes except last one
df_with_total.drop(index={0,1,2,3,4,5},inplace=True)
# renaming last row
df_with_total.rename(index={6:0}, inplace=True)
```

```python
# We continue with the rest months
```

```python
ma1 = ex6.iloc[:,0].str.startswith("Jun")
ex6_filt1 = ex6[ma1]
df_with_total1= add_row_with_column_sums(ex6_filt1)
df_with_total1.drop(index={0,1,2,3,4,5},inplace=True)
df_with_total1.rename(index={6:1}, inplace=True)
df_with_total1
```

```python
ma2 = ex6.iloc[:,0].str.startswith("Sep")
ex6_filt2 = ex6[ma2]
df_with_total2= add_row_with_column_sums(ex6_filt2)
df_with_total2.drop(index={0,1,2,3,4},inplace=True)
# 5 instead of 6 because there are not data for Sep of 2016
df_with_total2.rename(index={5:2}, inplace=True)
```

```python
ma3 = ex6.iloc[:,0].str.startswith("Dec")
ex6_filt3 = ex6[ma3]
df_with_total3= add_row_with_column_sums(ex6_filt3)
df_with_total3.drop(index={0,1,2,3,4},inplace=True)
# 5 instead of 6 because there are not data for Dec of 2016
df_with_total3.rename(index={5:3}, inplace=True)
```

```python
total_table= pd.concat([df_with_total, df_with_total1, df_with_total2, df_with_total3],axis=0)
# set new values in column "Date"
total_table.iloc[0:1,0:1]= "March"
total_table.iloc[1:2,0:1]= "June"
total_table.iloc[2:3,0:1]= "September"
total_table.iloc[3:4,0:1]= "December"
total_table
```

```python
fig, axis= plt.subplots(2,2)
```

```python
# Creating a 2x2 subplot and filling it with different charts

fig, axis= plt.subplots(2,2, figsize=(75,20))
# getting the "Date" column
x= ex2.iloc[:,0]
# getting the "Earnings on investment" column
y= ex2.iloc[:,1]
# plotting 
axis[0,0].plot(x,y)
axis[0,0].set_title("Earnings on investments per Time")
axis[0,0].set(xlabel="Years", ylabel="Earnings on investments")
# set x axis dates from 2011 to 2016
axis[0,0].set_xlim([ex2.iloc[21,0]],[ex2.iloc[0,0]])

x= ex2.iloc[:,0]
y= ex2.loc[:,"Contributions"]
axis[0,1].plot(x,y, label=("Total contributions","Employee contributions","Government contributions"))
axis[0,1].set_title("Contributions per Time")
axis[0,1].set(xlabel="Years", ylabel="Contributions")
axis[0,1].set_xlim([ex2.iloc[21,0]],[ex2.iloc[0,0]])
axis[0,1].legend()

x= ex2.iloc[:,0]
y= ex2.loc[:,"Payments"]
axis[1,0].plot(x,y, label=("Total payments","Benefits","Withdrawals"))
axis[1,0].set_title("Payments per Time")
axis[1,0].set(xlabel="Years", ylabel="Payments")
axis[1,0].set_xlim([ex2.iloc[21,0]],[ex2.iloc[0,0]])
axis[1,0].legend()

x= total_table.iloc[:,0]
# get the length of "Date" column
x_axis= np.arange(len(total_table.iloc[:,0]))
# move bars -0.2 and -0.4 units away from the x-axis to avoid overlapping
axis[1,1].bar(x_axis-0.2,total_table.iloc[:,1], width=0.2, color="b", align="center", label="Earnings on investments")
axis[1,1].bar(x_axis-0.4,total_table.iloc[:,2], width=0.2, color="g", align="center", label="Total contributions")
axis[1,1].bar(x_axis,total_table.iloc[:,5], width=0.2, color="r", align="center", label="Total payments")
axis[1,1].set_title("Sum of Quarters from 2011 to 2016")
axis[1,1].set(xlabel="Months", ylabel="Earnings on investments")
axis[1,1].legend()
# print the month names as x-axis indexing
plt.xticks(x_axis,total_table.iloc[:,0])
plt.show()
# the graph shows the sum of earnings on investments, total contributions and total payments for each month quarter from 2011 to 2016
# double click to zoom
```
